## Dummy Variables

---
Before I can split my data into training and testing datasets, I must convert my categorical features into binary dummy variables so that they can be fed into the model as numeric features. The machine learning models require numeric input so this must be done. To avoid multicollinearity, I will drop one of the dummies for each categorical feature.

The categorical features are: 'term', 'emp_length', 'home_ownership', 'verification_status', 'purpose', 'addr_state', 'initial_list_status', 'disbursement_method', and 'job'.

---

```python
# This code will create dummies from a categorical feature and then concatonate them to the main dataframe.
# Term.
df = pd.concat([df, pd.get_dummies(df.term, drop_first=True, prefix='term')], axis=1)
# I also need to drop the original feature from the dataframe.
df.drop('term', axis=1, inplace=True)
```

```python
# emp_length.
df = pd.concat([df, pd.get_dummies(df.emp_length, drop_first=True, prefix='emp_length')], axis=1)
df.drop('emp_length', axis=1, inplace=True)
```

```python
# home_ownership.
df = pd.concat([df, pd.get_dummies(df.home_ownership, drop_first=True, prefix='home_ownership')], axis=1)
df.drop('home_ownership', axis=1, inplace=True)
```

```python
# verification_status.
df = pd.concat([df, pd.get_dummies(df.verification_status, drop_first=True, prefix='verification_status')], axis=1)
df.drop('verification_status', axis=1, inplace=True)
```

```python
# purpose.
df = pd.concat([df, pd.get_dummies(df.purpose, drop_first=True, prefix='purpose')], axis=1)
df.drop('purpose', axis=1, inplace=True)
```

```python
# addr_state.
df = pd.concat([df, pd.get_dummies(df.addr_state, drop_first=True, prefix='addr_state')], axis=1)
df.drop('addr_state', axis=1, inplace=True)
```

```python
# initial_list_status.
df = pd.concat([df, pd.get_dummies(df.initial_list_status, drop_first=True, prefix='initial_list_status')], axis=1)
df.drop('initial_list_status', axis=1, inplace=True)
```

```python
# disbursement_method.
df = pd.concat([df, pd.get_dummies(df.disbursement_method, drop_first=True, prefix='disbursement_method')], axis=1)
df.drop('disbursement_method', axis=1, inplace=True)
```

```python
# job.
df = pd.concat([df, pd.get_dummies(df.job, drop_first=True, prefix='job')], axis=1)
df.drop('job', axis=1, inplace=True) 
```

```python
df.shape
```

```python
list(df.columns)
```

---
I will rename some of the dummy columns to get rid of spaces in the column name.

---
I will now check for correlation again to make sure that none of the features or dummies are highly correlated with each other.

---

```python
corr = df.corr()
```

```python
# This function will search through all feature pairs and display the correlation, allowing me to see if any highly correlated features need to be removed. 
def corr_check(X):
        df_corr2 = pd.DataFrame([[(i,j),X.loc[i,j]] for i,j in list(itertools.combinations(X, 2))],columns=['pairs','corr'])    
        print(df_corr2.sort_values(by='corr',ascending=False))

corr_check(corr)
```

---
There are no feature pairs with a correlation greater than +-0.75 so I do not need to drop. I will drop the id column as I no longer need it for the model.

---

```python
df.drop('id', axis=1, inplace=True)
```

## Train / Test Split

---
Because I have 10 years worth of data, I will train my model on pre-2017 data and test it on 2017 data. I will need to extract year from the datetime column issue_d and then use this to split the data.

---

```python
SQL_STRING = '''

select * from "Model_Features"

'''

df = pd.read_sql(SQL_STRING, con=engine)
```

```python
df['year'] = df['issue_d'].dt.year
```

```python
df.year.value_counts()
```

```python
# Training data is pre-2017.
train = df[df['year'] < 2017]
# Testing data is from 2017.
test = df[df['year'] == 2017]
# Now drop date columns.
train.drop(['issue_d', 'year'], axis=1, inplace=True)
test.drop(['issue_d', 'year'], axis=1, inplace=True)
```

```python
train.shape
```

```python
test.shape
```

---
I have 770,730 records to train the model on and a test set of 40,751 records. With all the dummies, there are 125 features (plus the target).

---

Now I need to split the target (loan_status) from the features to create target and feature variables.

---

```python
# Create target variables and remove target from predictors dataframe
train_target = train['loan_status'].values
train.drop(['loan_status'], axis=1, inplace=True)
test_target = test['loan_status'].values
test.drop(['loan_status'], axis=1, inplace=True)
```

----
Because the different features in my dataset have very different scales, e.g. annual income (10,000s - 100,000s), interest rates (%/100), etc. the data needs to be scaled so that these differences in scale do not have adverse effects on the model.

I will scale the data by standardising it. This will set the means of all features to 0 and all will have a standard deviation of 1.

---

```python
# Initialize the Scaler
ss = StandardScaler()
# Standardise the training set.
train_std = ss.fit_transform(train.values)
# Standardise the test data with train mean and standard deviation.
test_std = ss.transform(test.values)
```

```python
# Create X and y arrays for training and testing.
X_train = train_std     
y_train = train_target 
X_test = test_std
y_test = test_target
```

```python
print('X_train shape:', X_train.shape)
print('y_train shape:', y_train.shape)
print('X_test shape:', X_test.shape)
print('y_test shape:', y_test.shape)
```

```python
# Check the baseline accuracy.
baseline_acc = 1 - df.loan_status.mean()
baseline_acc
```

```python
# Check the imbalance of the test data set.
test_imbalance = 1 - y_test.mean()
test_imbalance
```

## Logistic Regression

---
The first model I will try is logistic regression, a good model for predicting binary targets.

---

```python
# Create logistic regression model and fit to X and y.
logreg1 = LogisticRegression()
logreg1.fit(X_train, y_train)
```

```python
# Check model score.
logreg1.score(X_train, y_train)
```

---
The score is approximately baseline.

---

```python
# Check cross validated score on training data.
cv_scores = cross_val_score(logreg1, X_train, y_train, cv=5)   # 5 folds.
print ("Cross-validated scores:", cv_scores)
print ('Mean:', cv_scores.mean())
```

---
Using cross validation, I can see that the model is performing consistently across all folds.

---

```python
logreg1.score(X_test, y_test)
```

```python
# Check cross validated score on test data.
cv_scores = cross_val_score(logreg1, X_test, y_test, cv=5)  # Again 5 folds.
print ("Cross-validated scores:", cv_scores)
print ('Mean:', cv_scores.mean())
```

---
The model also performs consistently across folds of the test set.

---

```python
# Confusion matrix to show false (-ve/+ve) and true (-ve/+ve)
print (confusion_matrix(y_test, logreg1.predict(X_test)))  # True (-ve),  False (+ve)
                                                           # False (-ve), True (+ve)
print ('\n')
# Classification report to show precision and recall.
print(classification_report(y_test, logreg1.predict(X_test), target_names=['Fully Paid', 'Default']))
```

---
This model has a lot more false negatives than false positives and not too many true positives. Because I want a model to be conservative when predicting loan defaults by having more false positives than false negatives, I want to maximise recall. Looking at the average/total recall score is misleading because it has a high score of 0.85. However, when looking at the recall scores for each class, fully Paid has a high recall score but default has a low recall score. Therefore, when looking at recall scores of the model, I want the default recall score to be as high as possible and closer to the recall score of fully paid.

---

```python
# Check coefficients and intercepts.
print('Logreg intercept:', logreg1.intercept_)
```

```python
# Create dataframe to look at coefficients of each feature.
logreg1_coefs = pd.DataFrame({'feature':train.columns,
                            'coef':logreg1.coef_.flatten(),     # Use .flatten() to convert coefs array into a 1 dimensional array.
                            'abs_coef':np.abs(logreg1.coef_.flatten())})

logreg1_coefs.sort_values('abs_coef', inplace=True, ascending=False)
logreg1_coefs.head(10)
```

---
Interest rate has by far the largest affect on the loan status. 60 month terms, debt-to-income ratio, and home rental also have high coefficients and are having a larger affect on the target.

---

```python
# Create ROC-AUC plot to compare true positive rate against false positive rate.
sns.set_style('darkgrid')

fig, ax = plt.subplots(figsize=(12,7))
# True/false positive rates calculated using predicted probabilities of test set.
fpr, tpr, _ = roc_curve(y_test,  logreg1.predict_proba(X_test)[::,1])   
# Calculate roc_auc from true/false positive rate
roc_auc = auc(fpr, tpr)
# Plot ROC-AUC
plt.plot(fpr, tpr, label='AUC = %0.5f' %(roc_auc), linewidth=4, color='#9f2305')
plt.title('ROC-AUC', size = 20)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate', fontsize=18)
plt.ylabel('True Positive Rate', fontsize=18)
plt.legend(loc=4, fontsize = 'xx-large')
plt.show()
```

---
Plotting the ROC curve shows that it has an area-under-the-curve of 0.70. This is not terrible but it could be improved.

### Gridsearch

---
To try and improve the logistic regression model, I will use a gridsearch to search for the best parameters.

---

```python
# Gridsearch parameters.
gs_params = {
    'penalty':['l1', 'l2'],    # Lasso or ridge regularisation.
    'solver':['liblinear'],
    'C':np.logspace(-5,5,10)   # C penalty, 1/(regularisation alpha/lambda). Lower score = larger penalty.
}

lr_gridsearch = GridSearchCV(LogisticRegression(), gs_params, cv=5, verbose=1, n_jobs=-1)
```

## Resampling

---
My models are having trouble making good predictions. One reason for this could be due to the high class imbalance whereby most loans do not default. I will try resampling the data to create a more similar ratio between defaults and fully paid loans. I will try both upsampling and downsampling the data and will create a 60:40 ratio for both, majority fully paid.

---

### Train / Test Split

---
I first need to do a train/test split. I can then resample the training set and leave the test set untouched. If I resample the whole dataset, then the test set could end up with records that are also in the training set. In order to properly evaluate the performance of a model, it must be tested on unseen data.

---

```python
SQL_STRING = '''

select * from "Model_Features"

'''

df = pd.read_sql(SQL_STRING, con=engine)
```

```python
df['year'] = df['issue_d'].dt.year
```

```python
df.year.value_counts()
```

```python
train = df[df['year'] < 2017]
test = df[df['year'] == 2017]
train.drop(['issue_d', 'year'], axis=1, inplace=True)
test.drop(['issue_d', 'year'], axis=1, inplace=True)
```

```python
train.shape
```

```python
test.shape
```

```python
train.loan_status.value_counts()
```

### Upsampling

---
I will start by upsampling the data. This upsamples the defaulted loans so that there are a higher proportion of defaults in the training set. The resampling is with replacement meaning there will be multiple copies of defaulted loans in the upsampled set in order to create a higher proportion of defaults.

---

First I will split the training set into one dataframe of fully paid loans and another dataframe of defaulted loans. Then I will upsample the defaulted dataframe and concatonate it back to the fully paid loans. This will leave me with an approximate ratio of 60:40, majority fully paid.

---

```python
# Create dataframe for fully paid loans.
train0 = train[train.loan_status==0]
```

```python
train0.shape
```

```python
# Create dataframe for defaulted loans.
train1 = train[train.loan_status==1]
```

```python
train1.shape
```

```python
# Upsample defaults, approx 60:40 (majority fully paid)
upsample = resample(train1[train.loan_status == 1], replace = True, n_samples = 400000, random_state = 0)
```

```python
# Concatonate fully paid loans and upsampled defaults.
train_up = pd.concat([train0, upsample], axis=0)
```
